```python
df = tb.copy()

df = df[df['p/l'] > 0]
df = df[df['p/vp'] > 0]
df = df[df['liquidez'] > 0]

df['graham'] = np.sqrt(22.5 / (df['p/l'] * df['p/vp'])) * df['preco']
df['desconto'] = df['preco'] / df['graham']
df.head()
```
