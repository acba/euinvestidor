```python
# encoding=utf8

import pandas as pd
import numpy as np

url = 'http://www.fundamentus.com.br/resultado.php'
tb = pd.read_html(url, flavor='bs4', decimal=",", thousands=".")[0]


# Tratando dados
tb['Div.Yield'] = pd.to_numeric(tb['Div.Yield'].str.replace('.', '').str.replace('%','').str.replace(',','.'))

tb['Mrg Ebit'] = pd.to_numeric(tb['Mrg Ebit'].str.replace('.', '').str.replace('%','').str.replace(',','.'))
tb['Mrg. Líq.'] = pd.to_numeric(tb['Mrg. Líq.'].str.replace('.', '').str.replace('%','').str.replace(',','.'))

tb['ROIC'] = pd.to_numeric(tb['ROIC'].str.replace('.', '').str.replace('%','').str.replace(',','.'))
tb['ROE'] = pd.to_numeric(tb['ROE'].str.replace('.', '').str.replace('%','').str.replace(',','.'))

tb['Cresc. Rec.5a'] = pd.to_numeric(tb['Cresc. Rec.5a'].str.replace('.', '').str.replace('%','').str.replace(',','.'))
```

```python
tb = tb.loc[tb['Liq.2meses'] > 0]

tb.head(10)
```

```python
tb = tb.loc[tb['Liq.2meses'] > 0]

## Analise EV/EBIT -> ROIC

#analise
tb = tb.sort_values(by=['EV/EBIT'])
tb['rank_ev'] = pd.Series(np.arange(tb.shape[0]), index=tb.index)

tb = tb.sort_values(by=['ROIC'], ascending=False)
tb['rank_roic'] = pd.Series(np.arange(tb.shape[0]), index=tb.index)

tb.loc[tb['EV/EBIT'] > 0].head()
```

```python
tb['rank_ev_roic'] = tb['rank_ev'] + tb['rank_roic']
tb.sort_values(by=['rank_ev_roic']).head(20)
```

```python
tb = tb.loc[tb['P/L'] * tb['P/VP'] < 25]

tb = tb.sort_values(by=['EV/EBIT'], ascending=False)
tb['rank_ev'] = pd.Series(np.arange(tb.shape[0]), index=tb.index)

tb = tb.sort_values(by=['ROIC'], ascending=False)
tb['rank_roic'] = pd.Series(np.arange(tb.shape[0]), index=tb.index)

tb['rank_ev_roic'] = tb['rank_ev'] + tb['rank_roic']
tb.sort_values(by=['rank_ev_roic']).head(20)
```
